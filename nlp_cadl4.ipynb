```python
!pip install numpy==2.0.2 scipy==1.14.1
!pip install gensim pyLDAvis nltk

import nltk
nltk.download("punkt")
nltk.download("punkt_tab")   # required in new NLTK
nltk.download("stopwords")

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

import gensim
from gensim import corpora
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

corpus = [
    "The economy is working better than ever according to the latest reports.",
    "The new movie released last weekend was a big hit among young audiences.",
    "Artificial Intelligence and Machine Learning are transforming industries.",
    "Political tensions are rising due to the upcoming elections.",
    "Healthcare research has advanced significantly in the last decade.",
    "The sports team won the championship after a thrilling final match.",
    "New technologies in renewable energy are shaping the future of sustainability.",
    "Academic papers are focusing more on data-driven research and analytics."
]

print("Original Corpus:")
for doc in corpus:
    print("-", doc)

stop_words = set(stopwords.words("english"))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [w for w in tokens if w.isalpha() and w not in stop_words]
    return tokens

processed_corpus = [preprocess(doc) for doc in corpus]

print("\nPreprocessed Corpus:")
for tokens in processed_corpus:
    print(tokens)

dictionary = corpora.Dictionary(processed_corpus)
doc_term_matrix = [dictionary.doc2bow(text) for text in processed_corpus]

lda_model = gensim.models.LdaModel(
    corpus=doc_term_matrix,
    id2word=dictionary,
    num_topics=3,
    random_state=42,
    passes=15,
    alpha="auto"
)

print("\n--- LDA Topics ---\n")
topics = lda_model.print_topics(num_words=5)
for idx, topic in topics:
    print(f"Topic {idx}: {topic}")

pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
